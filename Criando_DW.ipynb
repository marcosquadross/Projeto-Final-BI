```python
df_keys = {
    "2018" : df_18,
    "2019" : df_19,
    "2020" : df_20,
    "2021" : df_21,
    "2022" : df_22,
    "2023" : df_23
}
```

```python
for key, item in df_keys.items():
  item.to_sql(f"mortalidade_{key}", conn, if_exists="replace", index=False)
  print(f"Tabela mortalidade_{key} criada com sucesso! Com {len(item)} registros.")
```

```python
cursor = conn.cursor()
```

```python
#Nome de todas as tabelas
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tabelas = [t[0] for t in cursor.fetchall()]
print("Tabelas:", tabelas)
```

```python
colunas_por_tabela = {}

for tabela in tabelas:
    cursor.execute(f'PRAGMA table_info("{tabela}")')
    colunas = [row[1].upper() for row in cursor.fetchall()]  # pega os nomes em mai√∫sculo
    colunas_por_tabela[tabela] = colunas
```

```python
for item in colunas_por_tabela.values():
  print(item)
```
